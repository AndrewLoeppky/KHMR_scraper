```python
import pandas as pd
import requests
import bs4
```

```python
the_url = 'http://kickinghorseresort.com/conditions/advanced-weather-data/'
```

```python
page = requests.get(the_url)

#pass it on to beautifulsoup 
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, 'html.parser')

#Depending on how you want to find, you can use  findbyclass, findbytag, and #other methods 
the_soup = soup.findAll(class_="znColumnElement-innerContent")[1]
```

```python
the_table = str(the_soup).split("\r\n")[7:31]
the_table
```

```python
pd.read_csv(the_table, names=["Date", "Time", "Air Temp", "Wind Run", "Wind Speed", "Wind Dir", "Wind Gust", "Batt Volts"], sep='    ')
```
